```python
def train_model():
    data = "faces"
    path = os.listdir(data)
    faces = []
    ids = []
    img = path[0]
#     print(path)
    for image in path:
        image = "faces/"+image
#         print(image)
        img = Image.open(image).convert('L')
        imgNp = np.array(img,'uint8')
        id = int(os.path.split(image)[1].split(".")[1])
        
        faces.append(imgNp)
        ids.append(id)
    
    ids = np.array(ids)
    
    clf = cv2.face.LBPHFaceRecognizer_create()
    clf.train(faces,ids)
    clf.write("classifier.xml")
    messagebox.showinfo('Message','User Registered Successfully...')
```

```python
r_2 = tk.Tk()
r_2.title("Search attendence practice")
r_2.geometry("800x500+100+100")

def recapture():
    if(idEntry.get() == ""):
        messagebox.showinfo('Fill ID','Please enter ID...')
    else:
        id = idEntry.get()
        face_classifier = cv2.CascadeClassifier('haar-cascade-files-master/haarcascade_frontalface_default.xml')
        def face_cropped(img):
            gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
            faces = face_classifier.detectMultiScale(gray,1.3,5)
            cropped_face = img
            if faces is ():
                return None
            for(x,y,w,h) in  faces:
                cropped_face = img[y:y+h,x:x+w]
            return cropped_face

        count = 0
        cap = cv2.VideoCapture(0)
        temp = 1
        img_id = 0
        while True:
            ret , frame = cap.read()
            if face_cropped(frame) is not None:
                img_id += 1
                face = cv2.resize(face_cropped(frame),(200,200))
                face = cv2.cvtColor(face,cv2.COLOR_BGR2GRAY)
                path = "faces/user."+id+"."+str(img_id)+".jpg"
                cv2.imwrite(path,face)
                messagebox.showinfo('Message',"Camera Shut off Automatically at count 80...")
                cv2.putText(face,str(img_id),(50,50),cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),2)
                cv2.imshow("Cropped Face",face)
                if cv2.waitKey(1) == 13 or count == 80:
                    break
                count =  count + 1  
        cap.release()
        cv2.destroyAllWindows()
        train_model()

idLable = Label(r_2,text="Id",font=("Helvetica",20,"bold"),bg="#F3E3AE",fg="#383636")
idLable.pack()
idEntry  = Entry(r_2,width=30,bd=3)
idEntry.pack()

btn_recapture = tk.Button(r_2,text="Re-Capture",command=recapture).pack()

r_2.mainloop()
```
